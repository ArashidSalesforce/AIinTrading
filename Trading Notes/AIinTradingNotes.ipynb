```python
Comparing Tests for Normality:
- Normal Distributions are symmetirc and do not have fat tails or kurtosis. 
Box-whisker plots helps check if a distribution is symmetric or skewed (has a fat 
tail).
- QQ help us compare any two distributions, if two distributions are the same the 
QQ plot produces a fairly straight line.
- There are 3 tests to help us find if data distribution is normal: Shapiro-Wilk 
Test, D'Agostino-Pearson, and the Kolmogorov-Smirnov test. Each of these produce
p-value. 
- Shapiro-Wilk tends to perform better in a broader set of cases compared to
the D'Agostino-Pearson test. In part because the D'Agostino-Pearson test is used to
look for skewness, it cannot find non-normality in some odd non-normal distributions.
- The Kolmogorov-Smirnov test can be used to compare distributions other than
normal distributions, so it's similar to the QQ plot in its generality. To do a 
normality test, we would first rescale the data distribution (subtract the mean
and divide by its standard deviation), then compare the rescaled data distribution 
with the standard normal distribution (which has a mean of 0 and standard deviation
of 1). In general, the Shapiro-Wilk test tends to be a better test than the 
Kolmogorov-Smirnov test, but not in all cases.
- For complete thoroughness, you can use all the tests. Best to sanity check with
histograms, box-whisker plots, and QQ plots.
```

```python
Seasonal Adjustments using ARIMA:
- Industries have periods of time during the year where revenue can increase or 
decrease. Similarily investors and funds have unrealized gains or losses from stock
price increases or decrease respectively. The become realized when they sell, 
investors may sell stocks with loses before the end of the year, which may potentially
reduce their taxes and buy them back at the start of the year, this is called the 
"Jaunuary Effect". Removing seasonal effect can help to make the resulting time 
series stationary, and therefore more useful when feeding into an autoregressive 
moving average model. To remvoe seasonality, we can take the difference between each
data point one year prior. We'll refer to this as seasonal difference." For instance,
if you have monthly data, take the difference between August 2018 and August 2017 
and do the same for the rest of your data. It's common to take the "first difference"
either before or after taking the seasonal difference. If we took the "first 
difference" from the original time series, this would be taking August 2018 and 
subtracting July 2018. Next, to take the seasonal difference of the first difference,
this would mean taking the difference between (August 2018 - July 2018) and (August 
2017 - July 2017). You can check if the resulting time series is stationary, and
if so, run this stationary series through an autoregressive moving average model.

```

```python
Drift and Volatility Model and Mean Reversion:
- The drift and volatility model is also called a Brownian Motion model
and is a type of stochastic (random) volatility model. Stochastic volatility 
models are fundamental building blocks for estimating the price of options 
(call, puts, swaps), bonds. Before creating a model for example a option, we 
would first want a model for the underlying asset, the stock. That is what the
Brownian Motion Model attempts to describe. Stochastic volatility models were
used to improve upon the work of Black, Scholes and Merton, who came up with 
the first formula for pricing options. In relation to mean reversion, drift
and volatility model is a way to describe phenomena that we observe with 
stock prices. The model assumes there is a constant drift term with some
added randomness, so we can expect that a series will bounce around, but
still revert back to its long-term mean.
```

```python
Cointegation:
- To see if two stock time series are cointegrated, some linear combination
of their time series must form a stationary series.

Hedge Ratio:
- Hedge Ratio is the regression coefficient, for example the dependent variable
is stock2 and the independent variable is stock1.

Augmented Dickey Fuller Test:
- to check if two series are cointegrated, we can use the Augmented Dickey Fuller
Test (ADF). It is trying to determine if the linear combination of two series (
which is also a time series) is stationary. A series is stationary when its mean
covariance are constant, and also when the autocorrelation between one time period
and another only depends on the time duration between them, and not the specific
point in time of each observation.

                                         
```
